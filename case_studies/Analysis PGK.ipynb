## Import  Models for Analysis
To assist in the export/import process, some code and instructions have been provided for you. **Please keep in mind that the code assumes that the model objects will be exported/imported from the same directory where this notebook exits.** 

The base and integrated model objects will be stored as variables ``BASELINE_MODEL`` AND ``INTEGRATED_MODEL``. respectively

### Set Model IDs
* Set the variable ``MY_BASELINE_MODEL_ID`` to a string of the baseline model's identifier in order to import the model to the current directory.
* Set the variable ``MY_INTEGRATED_MODEL_ID`` to a string of the integrated model's identifier in order to import the model to the current directory.

```python
# example: MY_BASELINE_MODEL_ID = "RBC_PFK_G6PDH2r"

MY_BASELINE_MODEL_ID = "RBC_PGI_PGK_GND"
MY_INTEGRATED_MODEL_ID = "RBC_PGI_PGK_GND"
#can add more model IDs for more models
```

### Import Models

```python
from mass.io import json

#add extra model IDS below
models = []
for model_id, model_type in zip([MY_BASELINE_MODEL_ID, MY_INTEGRATED_MODEL_ID], ["INTEGRATED", "INTEGRATED"]):
    filepath = "../models/{0}_{1}.json".format(model_id, model_type)
    models.append(json.load_json_model(filepath))
    print("Loaded {0} Model with ID: {1}".format(model_type, model_id))
Baseline_RBC, Modified_RBC = models

numerical_values = {
    "Keq_PGK_D": 1/0.000056,
    "Keq_PGK_A": 1 / (5*0.029),
    "Keq_PGK_5": 1 / 0.00374194,
    "Keq_PGK_6": 1 / 2.4,
    "Keq_PGK_7": 1 / 0.205,
    "Keq_PGK_8": 1.87,
    "Keq_PGK_9": 0.45}
# Update the model with the parameters
Modified_RBC.update_parameters(numerical_values)

#set variable to a string of the model's identifier
MODULE_1_ID = 'HEX1'
MODULE_2_ID ='ADK1'

models = []
for model_id, model_type in zip([MODULE_1_ID, MODULE_2_ID], ["HEX1", "ADK1"]):
    filepath = "../models/{0}_{1}.json".format(model_id, model_type)
    models.append(json.load_json_model(filepath))
    print("Loaded {0} Model with ID: {1}".format(model_type, model_id))
HEX1, ADK1 = models

Baseline_RBC_HEX1=HEX1.merge(Baseline_RBC, inplace=False)
Baseline_RBC_HEX1.remove_reactions([Baseline_RBC_HEX1.reactions.HEX1])
Baseline_RBC_HEX1_ADK1=ADK1.merge(Baseline_RBC_HEX1, inplace=False)
Baseline_RBC_HEX1_ADK1.remove_reactions([Baseline_RBC_HEX1_ADK1.reactions.ADK1])

Modified_RBC_HEX1=HEX1.merge(Modified_RBC, inplace=False)
Modified_RBC_HEX1.remove_reactions([Modified_RBC_HEX1.reactions.HEX1])
Modified_RBC_HEX1_ADK1=ADK1.merge(Modified_RBC_HEX1, inplace=False)
Modified_RBC_HEX1_ADK1.remove_reactions([Modified_RBC_HEX1_ADK1.reactions.ADK1])
```

```python
# Define helper functions to make aggregate solutions
def make_energy_charge_solution(conc_sol):
    conc_sol.make_aggregate_solution(
        aggregate_id="energy_charge",
        equation="(atp_c + 0.5 * adp_c) / (atp_c + adp_c + amp_c)",
        variables=["atp_c", "adp_c", "amp_c"], update=True)

def make_active_fraction_solution(model, conc_sol, enzyme):
    enzyme = model.enzyme_modules.get_by_id(enzyme)
    active = enzyme.enzyme_module_forms_categorized.get_by_id(
        enzyme.id + "_Active")
    conc_sol.make_aggregate_solution(
        aggregate_id="active_fraction",
        equation="({0}) / ({1})".format(
            " + ".join([e.id for e in active.members]),
             str(strip_time(enzyme.enzyme_concentration_total_equation))),
        variables=enzyme.enzyme_module_forms, update=True)
def make_redox_ratio_solution(conc_sol):
    conc_sol.make_aggregate_solution(
        aggregate_id="redox_ratio",
        equation="(nadph_c) / (nadph_c + nadp_c)",
        variables=["nadph_c", "nadp_c"], update=True)
```

```python
# Create list of model IDs
model_ids = [Baseline_RBC,
             Modified_RBC]
# Define perturbations
perturbations = [
    {"kf_ATPM": "kf_ATPM * 0.85"},
    {"kf_ATPM": "kf_ATPM * 1.50"}]
# Define colors and legend labels
colors = ["red", "blue"]
labels = ["15% decrease", "50% increase"]
title=['Healthy model','Mutated model']
xlim=[[0.75,0.95],[0.7,0.95]]
ylim=[[0.55,0.85],[0.23,0.52]]
# Create figure and flatten axes into a list
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 5))
axes = axes.flatten()

# Iterate through models and axes instances
for i, (model_id, ax) in enumerate(zip(model_ids, axes)):
    # Load model and simulation
    model = model_id
    sim = Simulation(model, verbose=True)
    
    # Get ID of enzyme module in model
    enzyme_id = 'PGK'
    # Ensure models are in steady state and make aggregate solutions
    conc_sol, flux_sol = sim.find_steady_state(
        model, strategy="simulate", update_values=True, tfinal=1e4)
    make_energy_charge_solution(conc_sol)
    make_active_fraction_solution(model, conc_sol, enzyme_id)

    # Plot steady state lines
    ax.plot([conc_sol["energy_charge"]] * 2, [-0.1, 1.1],
            color="grey", linestyle="--")
    ax.plot([-0.1, 1.1], [conc_sol["active_fraction"]] * 2,
            color="grey", linestyle="--")

    # Initialize legend arguments
    legend = None
    time_points_legend = None
    # Iterate through ATP perturbations
    for j, perturbation in enumerate(perturbations):
        # Simulate model with perturbation and make aggregate solutions
        conc_sol, flux_sol = sim.simulate(
            model, time=(0, 1000), perturbations=perturbation)
        make_energy_charge_solution(conc_sol)
        make_active_fraction_solution(model, conc_sol, enzyme_id)

        # Set legends on the middle plot only
        legend = (labels[j] + " in ATP utilization", "upper outside")
        time_points_legend = "lower outside"

        # Make phase portrait
        plot_phase_portrait(
            conc_sol, x="energy_charge", y="active_fraction",
            ax=ax, legend=legend, xlim=xlim[i],ylim=ylim[i] ,                # Set axes and legend
            xlabel="Energy Charge (E.C.)",         # Axes labels
            ylabel=enzyme_id + " Active Fraction ($f_{A}$)",
            title=(title[i], {"size": "x-large"}), # Title
            linestyle="--", color=colors[j],       # Line color and style
            annotate_time_points="endpoints",      # Annotate time points
            annotate_time_points_color="black",
            annotate_time_points_legend=time_points_legend)
# Iterate through models and axes instances
plt.savefig("./figures/PGK_Figure3"+".pdf")
```
